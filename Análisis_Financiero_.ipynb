 **ANALISIS FINANCIERO COMPARATIVO ENTRE BANK OF AMERICA, BANCO SANTANDER Y BBVA ESPAA, Ene 24-Jun 24**

# Python enfocado en finanzas
Los datos sern tomados de la API de Yahoo Finance

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
tickers = ['BAC', 'SAN', 'BBVA']
```

```python
df_valuation = yf.download(tickers, start="2023-12-31", end='2024-06-30')
```

```python
plt.figure(figsize=(15,8))
plt.xticks(fontsize=6)
plt.yticks(fontsize=10)
df_valuation.boxplot()
plt.title('Boxplot combinado')
plt.show()
```

```python
plt.figure(figsize=(8,4))
colors = ['blue', 'green', 'black']
plt.hist(df_valuation['Adj Close'], bins=10, color=colors)
plt.title('Precios de cierre')
plt.show()
```
